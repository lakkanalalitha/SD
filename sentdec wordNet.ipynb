WordNet is a lexical database for the English language, which was created by Princeton, and is part of the NLTK corpus.
Used  to find the meanings of words, synonyms, antonyms, and more

```
from nltk.corpus import wordnet
```

```
import nltk
nltk.download('wordnet')

syns = wordnet.synsets("program")

syns
```

An example of a synset:

```
#An example of a synset:

print(syns[0].name())
```

prints just word

```
# prints just word

print(syns[0].lemmas()[0].name())
```

Definition of that first synset

```
# Definition of that first synset:

print(syns[0].definition())
```

Examples of the word in use:

```
# Examples of the word in use:

print(syns[0].examples())
```

How might we findout synonyms and antonyms to a word? 

The lemmas will be synonyms, and then you can use .antonyms to find the antonyms to the lemmas. 

```
syn = wordnet.synsets("good")

syn
```

```
synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
      #print(lemma.name())
      if lemma.antonyms():
            #print(lemma.antonyms())
            antonyms.append(lemma.antonyms()[0].name())
            
print(set(synonyms))
print(set(antonyms))
```

As you can see, we got many more synonyms than antonyms, since we just looked up the antonym for the first lemma.

 we can also easily use WordNet to compare the similarity of two words and their tenses, by incorporating the Wu and Palmer method for semantic related-ness.
 

```
# Let's compare the noun of "ship" and "boat:"

w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('boat.n.01')
print(w1.wup_similarity(w2))

w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('car.n.01')
print(w1.wup_similarity(w2))

w1 = wordnet.synset('ship.n.01')
w2 = wordnet.synset('cat.n.01')
print(w1.wup_similarity(w2))
```
