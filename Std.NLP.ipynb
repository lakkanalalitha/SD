**What is NLP**

NLP is a field at the intersection of

Computer science
AI
Linguistics

**goal:** 

goal for computers is to process or understand natural language in order to 

1)Perform tasks like - making appointments, buying things

2)Question answering - Siri, google assistant, facebook

3)Fully understanding and representing the meaning of language (which is more difficult)

   Perfect language understanding is AI-complete
